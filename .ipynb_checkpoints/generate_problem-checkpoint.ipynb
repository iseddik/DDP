```python
class Perceptron:
    activations = {
        "sigmoid": lambda x: (1 / (1 + np.exp(-x)))
    }

    def __init__(self, input_shape, activation='sigmoid'):
        self.w = np.random.random((input_shape, 1))
        self.b = np.random.random((1,))
        self.activation = activation
        
    def output(self, input):
        return self.activations.get(self.activation)(input.dot(self.w) + self.b)

class Layer:
    def __init__(self, num_neuron, input_shape, activation):
        self.num_neuron = num_neuron
        self.perceptrons = np.array([Perceptron(input_shape, activation) for _ in range(num_neuron)])

    def output(self, input):
        out = np.array([per.output(input) for per in self.perceptrons])
        return out.reshape((out.shape[0],))

class Model:
    def __init__(self):
        self.layers = []
    def add_layer(self, num_neuron, input_shape=None, activation="sigmoid"):
        if input_shape == None: input_shape = self.layers[-1].num_neuron
        self.layers.append(Layer(num_neuron, input_shape, activation))
    def forward(self, x):
        out = x
        for layer in self.layers:
            out = layer.output(out.T)
        return out
```

```python
model = Model()
model.add_layer(2, X[0].shape[0], "sigmoid")
model.add_layer(1)
model.add_layer(1)
model.forward(X[0])
```
